```
import pandas as pd 

url = 'https://raw.githubusercontent.com/mfmarlonferrari/datasetsForML/main/spotify_musics_api.csv'
dados = pd.read_csv(url)
```

```
dados.dtypes
```

```
dados = dados.drop(['target','song_title','artist'], axis=1)
```

```
dados['loudness'].hist()
```

```
from sklearn.cluster import KMeans

modelo = KMeans(n_clusters = 5)
y = modelo.fit_predict(dados[['energy','loudness']])
```

```
dados['playlist'] = y
```

```
import seaborn as sns

ax = sns.scatterplot(x='energy', y='loudness', data=dados, hue='playlist')
```
